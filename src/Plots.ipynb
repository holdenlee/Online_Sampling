```python
import matplotlib.pyplot as plt
import plotnine as gg

bot = 0.4
merrs=[0.1233166 , 0.35576233, 0.08198022, 0.56174148, 0.08159984,0.08074088]
ma = [1.0-x-bot for x in merrs]
#[mala, online laplace, PG, SGLD, SAGA-LD, Prec SAGA-LD]
def reshuff(l):
    return [l[3],l[1],l[0],l[2],l[4]]

plt.title('Marginal accuracies')
xs = range(0,10,2)
plt.bar(xs, reshuff(ma[:5]),bottom=bot)
plt.xticks(xs, reshuff(['MALA', 'Online\nLaplace', 'Polya-Gamma', 'SGLD', 'SAGA-LD']))
plt.show()
print(reshuff([1.0-x for x in merrs]))
```
